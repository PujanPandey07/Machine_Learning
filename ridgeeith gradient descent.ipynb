```python
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
```

```python
x,y=load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
reg=SGDRegressor(penalty='l2',max_iter=500,eta0=0.01,learning_rate='constant',alpha=0.001)
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.linear_model import Ridge
r=Ridge(alpha=0.01,max_iter=500,solver='sparse_cg')
```

```python
r.fit(x_train,y_train)
```

```python
y1_pred=r.predict(x_test)
```

```python
r2_score(y_test,y1_pred)
```

```python
import numpy as np

class meraridgegd:
    def __init__(self, lr=0.01, epochs=500, alpha=0.01):
        self.lr = lr
        self.epochs = epochs
        self.alpha = alpha
        self.m = None
        self.b = None

    def fit(self, x_train, y_train):
        # Add bias term (column of 1s) to x_train
        x_train = np.insert(x_train, 0, 1, axis=1)

        # Initialize theta (weights), including bias
        theta = np.ones(x_train.shape[1])

        for i in range(self.epochs):
            # Gradient of cost function for Ridge Regression
            # Note: Do not regularize the bias term (theta[0])
            grad = (x_train.T @ (x_train @ theta - y_train)) + self.alpha * np.append(0, theta[1:])
            theta -= self.lr * grad

        # Extract weights and bias
        self.b = theta[0]
        self.m = theta[1:]
        print("Weights (m):", self.m)
        print("Bias (b):", self.b)

    def predict(self, x):
        return x @ self.m + self.b
```

```python
ri=meraridgegd(0.005,500,0.01)
```

```python
ri.fit(x_train,y_train)
```

```python
y1_pred=ri.predict(x_test)
```

```python
r2_score(y_test,y1_pred)
```
