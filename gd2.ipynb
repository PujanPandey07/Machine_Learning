```python
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

  

a,b=make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=40)
```

```python
import matplotlib.pyplot as plt

plt.scatter(a, b)
plt.xlabel('Cgpa')
plt.ylabel('pp')
plt.title('Scatter Plot')
plt.show()

```

```python
import numpy as np

class mygradientdescent:
    def __init__(self):
        self.m = 37.332 
        self.c = 20       

    def myfit(self, a, b):
        a = a.ravel()
        for i in range(1000):
            error = b - self.m * a - self.c

            lossb = -2 * np.sum(error)
            lossm = -2 * np.sum(a * error)

            self.c -= 0.0001 * lossb 
            self.m -= 0.0001 * lossm

        return self.c, self.m
```

```python
gd=mygradientdescent()
t,d=(gd.myfit(a,b))
print(t)
print(d)
```
