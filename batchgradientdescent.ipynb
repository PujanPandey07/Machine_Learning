## lets do Batch Gradient descent


in batch gradient descent we use all rows data to compute the appropriate weights and biases


```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split as tt
```

```python
li=LinearRegression()
```

```python
X,y=load_diabetes(return_X_y=True)
```

```python
print(X.shape)
print(y.shape)
```

```python
x_train,x_test,y_train,y_test=tt(X,y,test_size=0.2,random_state=5)
```

```python
li.fit(x_train,y_train)
```

```python
print(li.coef_)
print(li.intercept_)
```

```python
y_pred=li.predict(x_test)
```

```python
r2=r2_score(y_test,y_pred)
```

```python
print(r2)
print(x_train.shape[1])
```

```python
class batchgdescent:
    def __init__(self,epochs=100,lr_rate=0.1):
        self.epochs=epochs
        self.lr_rate=lr_rate
        self.coeff=None
        self.intercept=None

    def gdfit(self,x_train,y_train):
        self.intercept=0
        self.coeff=np.ones(x_train.shape[1])
        for i in range(self.epochs):
            y_hat=np.dot(x_train,self.coeff)+self.intercept
            intercept_der=-2*np.mean(y_train-y_hat)
            coeff_dr=-2*np.dot((y_train-y_hat),x_train)/x_train.shape[0]
            self.intercept=self.intercept-self.lr_rate*intercept_der
            self.coeff=self.coeff-self.lr_rate*coeff_dr
        print( self.coeff,self.intercept)
    def gdpredict(self,x_test):
        return np.dot(x_test,self.coeff)+self.intercept
    
    
        
```

```python
bgd=batchgdescent(15000,0.001)
bgd.gdfit(x_train,y_train)
```

```python
y_pred=bgd.gdpredict(x_test)
```

```python
r2=r2_score(y_test,y_pred)
```
