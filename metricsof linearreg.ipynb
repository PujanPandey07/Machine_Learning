```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('polep.csv')
```

```python
x=df.iloc[:,0:1]
```

```python
y=df.iloc[:,1:2]
```

```python
from sklearn.model_selection import train_test_split as tt
```

```python
x_train,x_test,y_train,y_test=tt(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as mb
```

```python
li=LinearRegression()
li.fit(x_train,y_train)
```

```python
li.predict(x_test.iloc[5].values.reshape(1,1))
```

```python
mb.scatter(df['cgpa'],df['package'])
mb.plot(x_train,li.predict(x_train),color='red')
mb.xlabel('CGPA')

mb.ylabel('Package in lakhs per annum')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
ypred=li.predict(x_test)
```

```python
print("MAE:",mean_absolute_error(y_test,ypred))
```

```python
print("MsE:",mean_squared_error(y_test,ypred))
```

```python
print("r2score:",r2_score(y_test,ypred))
```

```python
print("RMsE:",np.sqrt(mean_squared_error(y_test,ypred)))
```
