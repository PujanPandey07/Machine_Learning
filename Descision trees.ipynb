```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
x,y=make_classification(n_samples=500,n_features=5,n_informative=4,n_redundant=0,n_classes=4,n_clusters_per_class=2,random_state=415,hypercube=False,class_sep=20)
```

```python
from sklearn.model_selection import train_test_split as tt
```

```python
x_train,x_test,y_train,y_test=tt(x,y,test_size=0.2,random_state=45)
```

```python
dt=DecisionTreeClassifier(max_depth=2)
```

```python
dt.fit(x_train,y_train)
```

```python
ypred=dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
```

```python
accuracy_score(y_test,ypred)
```

```python
confusion_matrix(y_test,ypred)
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(dt)
```
