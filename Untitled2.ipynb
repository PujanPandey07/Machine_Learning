```python
from bs4 import BeautifulSoup

def minify_html(html):
    soup = BeautifulSoup(html, "html.parser")
    return soup.prettify().replace('\n', '').replace('  ', '')

html = "<html>    <body>   Hello World </body> </html>"
print(minify_html(html))
```
