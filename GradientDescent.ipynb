```python
from sklearn.datasets import make_regression
import numpy as np
import pandas as pd
```

```python
x,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=40,random_state=25)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

## first lets apply ols

```python
from sklearn.linear_model import LinearRegression
a=LinearRegression()
a.fit(x,y)
```

```python
a.coef_
```

```python
a.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,a.predict(x),color='red')
```

Now lets apply gradient descent with m=22.1835 and b=0

```python
y_pred=( a.coef_*x+0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,a.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='blue',label='b=0')
plt.legend()
plt.show()
```

```python
b = -100
m = 22.185
lr = 0.01

epochs = 100

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*x.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * x + b

  plt.plot(x,y_pred)

plt.scatter(x,y)
     
```

```python
import numpy as np

class mygradientdescent:
    def __init__(self):
        self.m = 43.0717  
        self.c = 0        

    def myfit(self, a, b_target_sum):
        a = a.ravel()
        n = len(a)
        for i in range(10000):
            
            b_pred = self.m * a + self.c
            
            loss = np.sum(b_pred) - b_target_sum
            
            
            grad_c = loss  
            
            self.c -= 0.01 * grad_c  

        return self.c


    
```

```python
gd=mygradientdescent()
print(gd.myfit(a,b))
```
